```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2
```

```python
plt.subplot(1,2,1) #nrows, ncolumns, plot number
plt.plot(x,y,'b--')

plt.subplot(1,2,2)
plt.plot(x,y,'g*-')
plt.show()
```

# OOP Methods

## By creating an empty figure

```python
# OOP gives you more control
fig = plt.figure() # Creates blank canvas
axes = fig.add_axes([0.0,0.0,1.0,1.0]) # left, bottom, width, height(0-1)
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
axes.plot(x,y)
axes2.plot(y,x)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')

axes2.set_ylabel('Y Label')
axes2.set_xlabel('X Label')
axes2.set_title('Title')
plt.show()
```

## By using Subplots in OOP

The axes is a list of items meaning we can iterate through it;

```python
fig, axes = plt.subplots(nrows = 1, ncols = 3)
for current_ax in axes[1:]:
    current_ax.plot(x,y, 'o')
axes[0].plot(y, x, 'ro') # Access items on the axis list
plt.tight_layout() # Makes sure the layout does not overlap
plt.show()
```

### Figure size and DPI

```python
# e.g. using figure
fig = plt.figure(figsize=(6,3),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.show()
```

```python
# e.g. using subplots
fig, axes = plt.subplots(figsize = (6,2), nrows = 1, ncols = 2)
axes[0].plot(x, y, 'ro')
plt.tight_layout()
plt.show()
```

## Adding a Legend

```python
# Adding a legend
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# Label the lines using the label argument
ax.plot(x,x**2, linewidth = 3, alpha = 0.5, label = 'X square')  #linewidth sets the width of the line, alpha shows the transparency
ax.plot(x,x**3,linewidth = 3, label = 'X cube') 
# Call the legend method
ax.legend(loc = 'best') # the loc argument places the legend in a certain position
plt.show()
```

# Arguments

`linecolor` or `c` sets the color of the line. <br>
`linewidth` sets the with of the plot line. Its default value is 1. <br>
`alpha` shows how transparent the plot line is. <br>
`linestyle` or `ls` styles the plot line e.g `--`, `-.`. <br>
`marker` emphasizes each point with a specified marker e.g. 'o', '*'. <br>
`markersize` or `ms` is used to set the size of the marker. <br>
`markeredgecolor` or `mec` sets the color of the edge of the marker. <br>
`markerfacecolor` or `mfc` sets the color inside the edge of the marker. <br><br>


`grid()` adds a grid to the plot and has several parameters like `axis`, `linestyle`, `linewidth`. <br>
`set_xlim()` method sets a limit for the x-plot while the `set_ylim()` method sets a limit for the y-plot.<br>

You can set the values of marker, line and color all together e.g.'o:r' gives you a round marker with dotted lines and a red color. You can also use the parameter `fmt` for this.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**3, 'o:r', markersize = '10', mec = 'blue')

ax.set_xlim([0,1])
ax.set_ylim([0,2])

plt.show()
```

# Bar Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = np.array(['Dano', 'Victor', 'Joshua', 'Etini'])
y = np.array([22, 21, 22, 22])
plt.bar(x,y, color = 'black', width = 0.5)
fig
```

# Histogram

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = np.random.normal(170,10,250)
plt.hist(x, color = 'red')
fig
```

# Pie Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

x = np.array([45, 35, 20])
mylabels = ['Dano', 'Victor', 'Joshua']
myexplode = [0.2,0.0,0.0] # makes one of the wedges stand out
mycolors = ['black', 'green', 'orange']

plt.pie(x, labels = mylabels, explode = myexplode, colors = mycolors)
plt.legend(title = 'win ratio')
fig
```
