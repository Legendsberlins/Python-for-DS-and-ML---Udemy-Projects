# Pandas Built-in Data Visualization

This involves ways to visualize data directly from pandas. This is used for quick analysis of data not basically used for advanced work or presentation.

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col = 0) # index_col = 0 specifies that the first column should be used as an index rather than the default 0,1,2.
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2
```

You can use pandas to create a histogram of a certain column in the DataFrame with different ways. This calls matplotlib making you to be able to use matplot libs arguments.

### Method 1:

```python
df1['A'].hist(bins = 30)
plt.show()
```

### Method 2:

```python
df1['A'].plot(kind = 'hist', bins = 30)
plt.show()
```

### Method 3

```python
df1['A'].plot.hist()
plt.show()
```

## Area Plot

```python
df2.plot.area(alpha = 0.5) # alpha argument controls the transparency
plt.show()
```

## Bar Plot

```python
df2.plot.bar(stacked = True)
plt.show()
```

## Scatter Plot

In the example below, a scatterplot is created for DataFrame1 using A and B columns with it colored based on column C. The color map is magma.

```python
df1.plot.scatter(x = 'A', y = 'B', c = 'C', cmap = 'magma')
plt.show()
```

## Box Plot

```python
df2.plot.box()
plt.show()
```

```python
df3 = pd.DataFrame(np.random.randn(1000,2), columns = ['a', 'b'])
```

```python
df3.plot.hexbin(x = 'a', y = 'b', gridsize = 25)
plt.show()
```

```python
df3.plot.kde()
plt.show()
```
