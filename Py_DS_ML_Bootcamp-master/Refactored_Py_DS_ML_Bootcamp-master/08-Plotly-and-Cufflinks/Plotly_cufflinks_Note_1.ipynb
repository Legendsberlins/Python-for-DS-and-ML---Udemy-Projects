# Plotly and Cufflinks

**Plotly** is an interactive visualization library while **Cufflinks** connects Plotly to Pandas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from plotly import __version__
```

```python
import cufflinks as cf
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot # To use plotly offline, import some plotly functions
```

```python
init_notebook_mode(connected = True) # Connects JS to notebook. Plotly connects Pandas and Python to an interactive JS library
```

```python
cf.go_offline() # ensures use of cufflinks offline
```

```python
df = pd.DataFrame(np.random.randn(100,4), columns = ['A', 'B', 'C', 'D'])
```

```python
df.head(6)
```

```python
df2 = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [32, 43, 50]})
df2
```

## iplot() on lineplot

```python
df.iplot()
plt.show()
```

## iplot() on scatterplot

```python
df.iplot(kind = 'scatter', x = 'A', y = 'B', mode = 'markers')
plt.show()
```

## iplot() on barplot

```python
df2.iplot(kind = 'bar', x = 'Category', y = 'Values')
plt.show()
```

You can also use aggregate functions e.g. count, sum to create plots. E.g. the program below uses the aggregate function, sum which totals each column.

```python
df.sum().iplot(kind = 'bar')
plt.show()
```

## iplot() on boxplot

```python
df.iplot(kind = 'box')
plt.show()
```

## iplot() on 3d Surface

```python
df3 = pd.DataFrame({'x': [1,2,3,4,5], 'y': [10,20,30,40,50], 'z': [100,200,300,400,500]})
df3
```

```python
df3.iplot(kind = 'surface', colorscale = 'rdylbu')
plt.show()
```

## iplot() on histogram

```python
df.iplot(kind = 'hist', bins = 25)
plt.show()
```

## iplot() on spread

```python
df3[['x', 'y']].iplot(kind = 'spread')
plt.show()
```

## iplot() on bubble

Bubble plot is similar to scatter plot but the size of the plot depends on a variable

```python
df3.iplot(kind = 'bubble', x = 'x', y = 'z', size = 'y')
plt.show()
```

## iplot() on scatter matrix

Scatter matrix is similar to pairplot. It creates a scatter matrix fof every plot it can.

```python
df.scatter_matrix()
```
