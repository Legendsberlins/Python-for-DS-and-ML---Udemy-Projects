# Kmeans Clustering

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs # Generate artificial data
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data[0] # Shows the data points 
```

```python
data[0].shape # 200 data points(rows), 2 features(columns)
```

```python
data[1] # Gets the cluster assigned to each plot
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1]) # c makes the plots differentiated by clusters
plt.show()
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_ # Returns the datapoints of the 4 centroids
```

### **N/B:** A centroid is the average of all data points in a cluster.

```python
kmeans.labels_ # Labels the model believes to be true
```

## Compare original with Models prediction

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
plt.show()
```
